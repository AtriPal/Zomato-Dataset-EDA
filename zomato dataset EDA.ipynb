```python
df_country = pd.read_excel('../input/zomato-eda/Country-Code.xlsx') 
df_country.head()
```

```python
final_data = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_data.head()
```

```python
## check datatypes

final_data.dtypes
```

```python
final_data.columns
```

```python
 country_names=final_data['Country'].value_counts().index
```

```python
country_val = final_data.Country.value_counts().values
```

```python
# Pie chart --> Top 3 countries
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records are from India, after that USA and then UK

```python
final_data.columns
```

```python
ratings = final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

 Observation
 1. when rating 4.5 to 4.9 --> Excellent
 2. when rating 4.0 to 4.4 --> Very good
 3. when rating 3.5 to 3.9 --> Good
 4. when rating 2.5 to 3.4 --> Average
 5. when rating 1.8 to 2.4 --> Poor

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Aggregate rating', y='Rating count', data = ratings, hue='Rating color', palette=['grey','red','orange','yellow','green','#053b07'])
```

observation :
1. not rated count is very high
2. maximum no of rating are between 2.5 to 3.4

```python
## count plot

sns.countplot(data=ratings, x='Rating color',palette=['grey','red','orange','yellow','green','#053b07'] )
```
